```python
# Initial imports
import requests
from dotenv import load_dotenv
import os 
import json
import pandas as pd
```

```python
load_dotenv()
```

```python
#Get API Key stored as variable
api_key = os.getenv("CENSUS_DATA_API_KEY")
```

```python
#Check that key was retrieved
type(api_key)
```

```python
# Define the base request URL
state_pop_url = "https://api.census.gov/data/2021/pep/population?get=NAME,POP_2021&for=state:*&key="

# Concatenate request_url and api_key. Store as new variable
request_url = state_pop_url + api_key
```

```python
state_pop_df = pd.read_json(request_url)
state_pop_df.head()
```

```python
state_pop_df.rename(columns=state_pop_df.iloc[0]).loc[1:]
```

```python
state_pop_df.drop(columns='state', inplace=True)
```

```python
state_pop_df.set_index(state_pop_df['state'])
```
